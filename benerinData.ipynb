```python
import pandas as pd

# 1. Daftar file tahunan beserta tahunnya
files_yearly = {
    'Pemeliharaan Kendaraan 2023.csv': 2023,
    'Pemeliharaan Kendaraan 2024.csv': 2024,
    'Pemeliharaan Kendaraan 2025.csv': 2025
}

# Membaca dan menggabungkan file 2023-2025
df_yearly_list = []
for f, tahun in files_yearly.items():
    temp_df = pd.read_csv(f, sep=';')
    temp_df['Tahun'] = tahun   # <-- Menambahkan kolom tahun sesuai file
    temp_df['Sumber'] = f 
    df_yearly_list.append(temp_df)

df_yearly = pd.concat(df_yearly_list, ignore_index=True)

# 2. Membaca data.csv
df_data = pd.read_csv('data.csv', sep=';')

# 3. Membersihkan kolom NILAI di data.csv
df_data['NILAI'] = df_data['NILAI'].str.replace('Rp ', '', regex=False).str.replace('.', '', regex=False).astype(float)

# 4. Melakukan Rename kolom data.csv
mapping = {
    'NOPOL': 'Nopol',
    'REKANAN': 'Vendor',
    'NILAI': 'Total Biaya',
    'JENIS PEMELIHARAAN': 'Keterangan',
    'MERK': 'Type',
    'KETERANGAN': 'Status Proses',
    'TANGGAL': 'Tanggal_Full'
}
df_data_renamed = df_data.rename(columns=mapping)

# 5. Menyamakan kolom 'Bulan' & 'Tahun' (Ekstrak dari Tanggal_Full)
df_data_renamed['Tanggal_Full'] = pd.to_datetime(df_data_renamed['Tanggal_Full'], dayfirst=True)
month_map = {
    1: 'Januari', 2: 'Februari', 3: 'Maret', 4: 'April', 5: 'Mei', 6: 'Juni',
    7: 'Juli', 8: 'Agustus', 9: 'September', 10: 'Oktober', 11: 'November', 12: 'Desember'
}
df_data_renamed['Bulan'] = df_data_renamed['Tanggal_Full'].dt.month.map(month_map)
df_data_renamed['Tahun'] = df_data_renamed['Tanggal_Full'].dt.year # <-- Menambahkan kolom tahun otomatis
df_data_renamed['Sumber'] = 'data.csv'

# 6. Menggabungkan data tahunan dengan data.csv
df_final = pd.concat([df_yearly, df_data_renamed], ignore_index=True)

print("Proses penggabungan selesai! Kolom 'Tahun' telah ditambahkan.")
```

```python
# 1. Pastikan kolom Total Biaya adalah tipe angka (numeric)
# Ini perlu dilakukan jika ada data yang masih terbaca sebagai teks dengan titik/koma
df_final['Total Biaya'] = pd.to_numeric(
    df_final['Total Biaya'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False), 
    errors='coerce'
).fillna(0)

# 2. Hapus baris yang Total Biayanya 0 atau 1
df_final = df_final[~df_final['Total Biaya'].isin([0, 1])]

# 3. (Opsional) Hapus juga jika ada nilai yang negatif atau di bawah 1
# df_final = df_final[df_final['Total Biaya'] > 1]

print(f"Pembersihan selesai! Sisa baris sekarang: {len(df_final)}")
df_final.head()
```

```python
# 1. Membuat 'Kamus' Referensi dari Nopol ke Jenis Kendaraan & Type
# Kita ambil baris yang datanya lengkap, lalu hapus duplikat Nopol-nya
referensi_jenis = df_final.dropna(subset=['Jenis Kendaraan']).drop_duplicates('Nopol')[['Nopol', 'Jenis Kendaraan']]
referensi_type = df_final.dropna(subset=['Type']).drop_duplicates('Nopol')[['Nopol', 'Type']]

# 2. Mengisi (Fillna) kolom yang kosong berdasarkan Nopol
# Kita jadikan Nopol sebagai index sementara agar pengisian lebih cepat
df_final = df_final.set_index('Nopol')

df_final['Jenis Kendaraan'] = df_final['Jenis Kendaraan'].fillna(referensi_jenis.set_index('Nopol')['Jenis Kendaraan'])
df_final['Type'] = df_final['Type'].fillna(referensi_type.set_index('Nopol')['Type'])

# Kembalikan Nopol menjadi kolom biasa
df_final = df_final.reset_index()

# 3. Jalankan pembersihan yang diminta sebelumnya
# Hapus Total Biaya 0 atau 1
df_final['Total Biaya'] = pd.to_numeric(df_final['Total Biaya'], errors='coerce').fillna(0)
df_final = df_final[~df_final['Total Biaya'].isin([0, 1])]

# Hapus Keterangan yang kosong
df_final = df_final.dropna(subset=['Keterangan'])

print("Selesai! Jenis Kendaraan dan Type telah diisi berdasarkan referensi Nopol.")
```

```python
import pandas as pd

# 1. Load data
df = pd.read_csv('Data_Kendaraan_Bersih.csv', sep=';')

# 2. Standarisasi teks (Huruf besar & hapus spasi)
df['Type'] = df['Type'].astype(str).str.strip().str.upper()

def pilih_tipe_terbaik(group):
    # Hilangkan pilihan 'OTHER', 'OTHERS', atau 'OTHER D' dari pencarian sementara
    specific_types = group[~group.isin(['OTHER', 'OTHERS', 'OTHER D', 'NAN', 'UNKNOWN'])]
    
    if not specific_types.empty:
        # Jika ada tipe spesifik, ambil yang paling sering muncul di antara tipe spesifik itu
        return specific_types.value_counts().idxmax()
    else:
        # Jika memang semuanya 'OTHER', barulah ambil 'OTHER'
        return group.value_counts().idxmax()

# 3. Buat referensi Nopol ke Type dengan logika "Anti-Other"
nopol_type_ref = df.groupby('Nopol')['Type'].apply(pilih_tipe_terbaik).reset_index()

# 4. Gabungkan kembali ke dataframe utama
df_final_clean = df.drop(columns=['Type']).merge(nopol_type_ref, on='Nopol', how='left')

# 5. Susun kolom dan simpan
cols = ['Nopol', 'Bulan', 'Vendor', 'Type', 'Total Biaya', 'Keterangan', 'Tahun', 'Sumber']
df_final_clean = df_final_clean[cols]

df_final_clean.to_csv('dataBener.csv', index=False, sep=';')

print("Pembersihan selesai! Nopol dengan tipe campuran kini menggunakan tipe yang lebih spesifik.")
```

```python
import pandas as pd

# 1. Load Data
df = pd.read_csv('dataBener.csv', sep=';')

# 2. Standarisasi Teks (Agar pencarian lebih akurat)
df['Type'] = df['Type'].astype(str).str.upper()
df['Vendor_Original'] = df['Vendor'] # Simpan nama asli buat cadangan

# 3. Fungsi Mapping Vendor berdasarkan Type
def ganti_vendor_by_type(row):
    v_type = row['Type']
    
    # Logika Mapping: Jika Type mengandung kata tertentu, arahkan ke Vendor Utama
    if any(x in v_type for x in ['INNOVA', 'HIACE', 'CAMRY', 'FORTUNER', 'KIJANG']):
        return 'TOYOTA SERVICE CENTER (ASTRA)'
    
    elif any(x in v_type for x in ['PANTHER', 'ISUZU', 'NMR', 'NKR']):
        return 'ISUZU SERVICE CENTER (ASTRA)'
    
    elif any(x in v_type for x in ['APV', 'SUZUKI']):
        return 'SUZUKI SERVICE CENTER (UMC)'
    
    elif 'HINO' in v_type:
        return 'HINO SERVICE CENTER (DUTA CEMERLANG)'
    
    elif 'MITSUBISHI' in v_type:
        return 'MITSUBISHI SERVICE CENTER'
    
    # Jika tidak ada yang cocok, gunakan Vendor asli yang sudah dibersihkan
    return row['Vendor']

# 4. Eksekusi Perubahan
df['Vendor'] = df.apply(ganti_vendor_by_type, axis=1)

# 5. Simpan ke File Baru
df.to_csv('Data_Pemeliharaan_Vendor_Mapped.csv', index=False, sep=';')

print("Berhasil! Vendor sekarang dikelompokkan berdasarkan Type kendaraan.")
print(df[['Type', 'Vendor']].drop_duplicates().head(10))
```

```python
import pandas as pd

# 1. Load Data
df = pd.read_csv('Data_Pemeliharaan_Vendor_Mapped.csv', sep=';')

# 2. Fungsi untuk mengelompokkan Keterangan
def kategori_kerusakan(text):
    text = str(text).upper()
    
    if 'BERAT' in text:
        return 'RUSAK BERAT'
    elif 'SEDANG' in text:
        return 'RUSAK SEDANG'
    elif 'RINGAN' in text:
        return 'RUSAK RINGAN'
    elif 'TUNE UP' in text or 'BERKALA' in text:
        return 'TUNE UP'
    else:
        # Jika tidak mengandung kata kunci di atas, tetap gunakan teks asli atau 'LAINNYA'
        return 'TUNE UP' # Biasanya perawatan rutin masuk ke Tune Up

# 3. Eksekusi Perbaikan
df['Keterangan'] = df['Keterangan'].apply(kategori_kerusakan)

# 4. Simpan ke CSV Final
df.to_csv('Data_Pemeliharaan_Vendor_Mapped.csv', index=False, sep=';')

print("Kolom Keterangan berhasil disederhanakan!")
print(df['Keterangan'].value_counts())
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Load Data
df = pd.read_csv('dataBener.csv', sep=';')

# Fungsi Preprocessing (Pembersihan Awal)
def preprocess_text(text):
    if pd.isna(text): return ""
    text = str(text).lower()
    # Hapus kata-kata badan usaha yang tidak unik
    stopwords = ['pt.', 'pt ', 'cv.', 'cv ', 'ud.', 'ud ', 'tbk.', 'tbk', '(', ')', '-']
    for word in stopwords:
        text = text.replace(word, ' ')
    text = re.sub(r'[^\w\s]', ' ', text) # Hapus sisa tanda baca
    text = re.sub(r'\s+', ' ', text).strip() # Hapus spasi ganda
    return text

# Siapkan data unik untuk dianalisa
vendor_counts = df['Vendor'].value_counts()
unique_vendors = vendor_counts.index.tolist()
cleaned_vendors = [preprocess_text(v) for v in unique_vendors]

# 2. Vektorisasi dengan TF-IDF & N-Grams
# ngram_range=(3,3) berarti kita memecah teks per 3 huruf untuk mendeteksi kemiripan ejaan
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(3, 3))
tfidf_matrix = vectorizer.fit_transform(cleaned_vendors)

# 3. Hitung Kemiripan (Cosine Similarity)
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# 4. Clustering (Pengelompokan Otomatis)
group_map = {} 
visited = set()
threshold = 0.55 # Ambang batas kemiripan (bisa diatur, 0.55 cukup moderat)

for i, vendor_i in enumerate(unique_vendors):
    if vendor_i in visited: continue
    
    # Vendor ini menjadi "Ketua Grup"
    group_label = vendor_i 
    group_map[vendor_i] = group_label
    visited.add(vendor_i)
    
    # Cari vendor lain yang mirip dengan Ketua Grup ini
    for j, vendor_j in enumerate(unique_vendors):
        if vendor_j in visited: continue
            
        similarity = cosine_sim[i, j]
        
        # Jika kemiripan di atas threshold, masukkan ke grup ini
        if similarity > threshold:
            group_map[vendor_j] = group_label
            visited.add(vendor_j)

# Terapkan hasil grouping ke DataFrame
df['Vendor_TextMining'] = df['Vendor'].map(group_map)

# Simpan dan Tampilkan Contoh Hasil
print("Contoh Hasil Pengelompokan Text Mining:")
print(df[['Vendor', 'Vendor_TextMining']].drop_duplicates().head(10))

df.to_csv('dataBener_TextMining.csv', index=False, sep=';')
```
